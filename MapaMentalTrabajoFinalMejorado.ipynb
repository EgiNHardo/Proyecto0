```python
# Paso 1: Instalar matplotlib (por si no está)
#   !pip install matplotlib

# Paso 2: Importar librerías
#   import matplotlib.pyplot as plt

# Paso 3: Crear la figura
fig, ax = plt.subplots(figsize=(14, 12))
ax.axis("off")

# Paso 4: Datos jerárquicos
secciones = {
    "Neurobiología de las emociones": [
        " Sistema límbico: amígdala, hipocampo, ínsula",
        " Corteza prefrontal: regulación emocional",
        " Eje HPA: estrés crónico, cortisol elevado",
        " Atrofia cortical y neuroinflamación"
    ],
    "Factores psicosociales": [
        " Presión académica extrema",
        " Aislamiento institucional",
        " Falta de red de apoyo"
    ],
    "Manifestaciones clínicas": [
        " Déficit cognitivo: memoria, atención",
        " Síntomas emocionales: ansiedad, fatiga",
        " Conductas evitativas: procrastinación, abandono"
    ],
    "Consecuencias académicas y personales": [
        " Desmotivación profunda",
        " Retiro de la universidad",
        " Fracaso académico"
    ]
}

# Paso 5: Colores para cada sección
colores = ["#B400B4", "#00F200", "#21C5FF", "#FFFF00"]
color_map = dict(zip(secciones.keys(), colores))

# Título
ax.text(0.5, 0.95, "Gestión emocional en educación superior", fontsize=21,
        ha="center", weight="light", color="darkblue")

# Posición vertical inicial
y = 0.85
line_height = 0.04

# Paso 6: Dibujar las secciones e ítems
for seccion, items in secciones.items():
    ax.text(0.05, y, f"{seccion}", fontsize=14, weight="bold", color=color_map[seccion])
    y -= line_height
    for item in items:
        ax.text(0.10, y, f"└── {item}", fontsize=12, color="black", family="monospace")
        y -= line_height
    y -= 0.02  # Espacio adicional entre secciones

# Paso 7: Guardar la imagen
plt.savefig("mapa_mental_llaves_jerarquico.png", dpi=300, bbox_inches='tight')
plt.show()
```
