Otra forma de subir el video

```python
from google.colab import files

uploaded = files.upload()

# Obtiene el nombre del archivo subido
import os
video_filename = list(uploaded.keys())[0]
print(f"Video cargado: {video_filename}")
```

Cargando WHISPER y transcribiendo el video

```python
import whisper

# Cargar el modelo base (puedes usar "small", "medium" o "large" si deseas mÃ¡s precisiÃ³n)
model = whisper.load_model("base")

# Transcribir directamente el video (Whisper usa ffmpeg para extraer el audio internamente)
result = model.transcribe(video_filename)

# Mostrar el texto transcrito
print("\nğŸ” TranscripciÃ³n:")
print(result["text"])
```

Guardando la transcripciÃ³n como archivo .txt para descargarlo como archivo

```python
from google.colab import files

# Guardar la transcripciÃ³n en un archivo de texto
with open("transcripcion.txt", "w", encoding="utf-8") as f:
    f.write(result["text"])

# Descargar el archivo de texto
files.download("transcripcion.txt")
```

```python
# 3. Verificar si hay GPU y usarla
import torch
import whisper

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"ğŸ–¥ï¸ Usando dispositivo: {device.upper()}")

# Cargar el modelo en el dispositivo adecuado
model = whisper.load_model("base", device=device)
```

```python
# 4. Transcribir el video
# Si no se ha definido el nombre del archivo, vuelve a asignarlo
import os
if 'video_filename' not in locals():
    video_filename = next((f for f in os.listdir() if f.lower().endswith(('.mp4', '.mov', '.avi'))), None)

if video_filename:
    result = model.transcribe(video_filename)
    print("ğŸ“ TranscripciÃ³n (vista previa):\n")
    print(result["text"][:1000])
else:
    print("âŒ No se encontrÃ³ ningÃºn archivo de video vÃ¡lido en el entorno.")
```

```python
# 5. Guardar y descargar como archivo .txt
from google.colab import files

with open("transcripcion.txt", "w", encoding="utf-8") as f:
    f.write(result["text"])

files.download("transcripcion.txt")
```
