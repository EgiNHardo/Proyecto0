```python
# Importar librer√≠as
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Instala dependencias si es necesario
#   !pip install networkx matplotlib

# Importar librer√≠as
#   import matplotlib.pyplot as plt
#   import networkx as nx

# Crear el grafo del mapa mental
G = nx.Graph()

# Nodo central
G.add_node("Gesti√≥n emocional en educaci√≥n superior")

# Subnodos principales
subnodes = [
    "Neurobiolog√≠a de las emociones",
    "Factores psicosociales",
    "Manifestaciones cl√≠nicas",
    "Consecuencias acad√©micas y personales"
]

for node in subnodes:
    G.add_node(node)
    G.add_edge("Gesti√≥n emocional en educaci√≥n superior", node)

# Subnodos de neurobiolog√≠a
neuro_nodes = [
    "Sistema l√≠mbico: am√≠gdala, hipocampo, √≠nsula",
    "Corteza prefrontal: regulaci√≥n emocional",
    "Eje HPA: estr√©s cr√≥nico, cortisol elevado",
    "Atrofia cortical y neuroinflamaci√≥n"
]
for node in neuro_nodes:
    G.add_node(node)
    G.add_edge("Neurobiolog√≠a de las emociones", node)

# Subnodos psicosociales
social_nodes = [
    "Presi√≥n acad√©mica extrema",
    "Aislamiento institucional",
    "Falta de red de apoyo"
]
for node in social_nodes:
    G.add_node(node)
    G.add_edge("Factores psicosociales", node)

# Subnodos de manifestaciones cl√≠nicas
manifest_nodes = [
    "D√©ficit cognitivo: memoria, atenci√≥n",
    "S√≠ntomas emocionales: ansiedad, fatiga",
    "Conductas evitativas: procrastinaci√≥n, abandono"
]
for node in manifest_nodes:
    G.add_node(node)
    G.add_edge("Manifestaciones cl√≠nicas", node)

# Subnodos de consecuencias acad√©micas
conseq_nodes = [
    "Desmotivaci√≥n profunda",
    "Retiro del posgrado",
    "Fracaso acad√©mico"
]
for node in conseq_nodes:
    G.add_node(node)
    G.add_edge("Consecuencias acad√©micas y personales", node)

# Posiciones y graficado
pos = nx.spring_layout(G, k=0.9, seed=42)
plt.figure(figsize=(15, 10))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightblue", font_size=9, font_weight="bold")
plt.title("Mapa mental: Etiolog√≠a y manifestaciones de los des√≥rdenes emocionales en el posgrado", fontsize=14)
plt.axis("off")
plt.savefig("mapa_mental_emociones.png", dpi=300, bbox_inches='tight')  # Guardar imagen
plt.show()
```
