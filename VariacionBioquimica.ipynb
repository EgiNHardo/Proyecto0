```python
# Importar librer칤as
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Instala dependencias si es necesario
#   !pip install pandas seaborn matplotlib

# Importar librer칤as
#   import pandas as pd
#   import seaborn as sns
#   import matplotlib.pyplot as plt

# Datos simulados
data = {
    "Estructura cerebral": ["Hipocampo", "Am칤gdala", "Corteza prefrontal", "칈nsula", "ACC"],
    "Cortisol": [-20, 10, -15, -5, -10],
    "Serotonina": [-15, -5, -10, -5, -5],
    "Dopamina": [-10, 5, -10, 0, -5],
    "Noradrenalina": [-10, 5, -5, 0, -5]
}

df = pd.DataFrame(data)
df_melt = df.melt(id_vars="Estructura cerebral", var_name="Neuroqu칤mico", value_name="Variaci칩n (%)")

# Crear gr치fico
plt.figure(figsize=(12, 6))
sns.barplot(data=df_melt, x="Estructura cerebral", y="Variaci칩n (%)", hue="Neuroqu칤mico", palette="viridis")
plt.title("Variaci칩n de compuestos bioqu칤micos por estructura cerebral (efectos del estr칠s cr칩nico)", fontsize=13)
plt.axhline(0, color='gray', linestyle='--')
plt.legend(title="Neuroqu칤mico")
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig("variacion_bioquimica_cerebro.png", dpi=300, bbox_inches='tight')  # Guardar imagen
plt.show()
```
